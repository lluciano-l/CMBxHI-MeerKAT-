```python
from scipy.fft import fft, ifft

def KLIM(X,XS): #lightcone evolution kernel for LIM
    ans = np.zeros_like(X) 
    i = np.where(X>XS)[0][0]
    ans[i:] = 0
    ans[:i] = (X[:i]-XS)**2
    return(ans)
# “more spectral line emission occurs as the universe gets older”

```
